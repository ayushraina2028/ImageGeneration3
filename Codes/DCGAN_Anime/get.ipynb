```python
import os
import pandas as pd
from PIL import Image
import io

# Define the path to the Parquet file and the output folder
parquet_file_path = 'train.parquet'
output_folder = 'output_images'

# Create the output directory if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Read the Parquet file
df = pd.read_parquet(parquet_file_path)

df
```

```python
df = df[["imageB"]]
```

```python
df
```

```python
# Read bytes from imageB column and save images to output folder
image = df.iloc[1]["imageB"]
```

```python
bytes = image['bytes']
```

```python
# Convert bytes to image
image = Image.open(io.BytesIO(bytes))

# Save image to output folder
image.save(os.path.join(output_folder, 'image.png'))
```

```python
for i in range(len(df)):
    image = df.iloc[i]["imageB"]
    bytes = image['bytes']
    image = Image.open(io.BytesIO(bytes))
    image.save(os.path.join(output_folder, f'image_{i}.png'))
    print(f"Image {i} saved")
```
